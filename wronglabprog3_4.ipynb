```python
# Задание task_03_04_01

# Выполнила: Александрова Екатерина.
# Группа: ЦИБ - 251

def load_data():
    """Вернуть список вещественных чисел, введенных с клавиатуры.

    Числа вводятся в одной строке.
    """

    s = input()
    values = list(map(float, s.split()))
    return values


def save_data(values, filename):
    """Записать данные в файл.

    Параметры:

    - values (список чисел с плавающей запятой): список вещественных чисел;
    - filename (str): имя файла.

    Функция не обрабатывает исключения.
    """

    with open(filename, 'w') as f:
        for x in values:
            f.write(str(x) + '\n')


try:
    filename = input()
    values = load_data()
    save_data(values, filename)

except ValueError:
    print("Числа введены с ошибкой.")

except (OSError, IOError) as err:
    print("Ошибка при сохранении файла с данными:", err)

except Exception as err:
    print("Произошла ошибка!")
    print("Тип:", type(err))
    print("Описание:", err)
```
